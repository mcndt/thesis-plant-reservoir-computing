### Datasets

```python
from src.model.rc_dataset import ExperimentDataset

DATASET_PATH = '../datasets/hydroshoot_large_trimmed.csv'

dataset = ExperimentDataset(csv_path=DATASET_PATH)
```

## Net photosynthesis


```python
from pipeline_hydroshoot import SingleReservoirGenerator


reservoir = SingleReservoirGenerator(state_var='state_Rg', run_ids=[0])
res_data = reservoir.transform((dataset,))
print(res_data.shape)


plot_data = res_data[0, :100]
sns.relplot(kind='line', data=plot_data, legend=False, alpha=0.3)
plt.title('State "Net photosynthesis rate"')
plt.show()
```

```python
reservoir = SingleReservoirGenerator(state_var='state_An', run_ids=[0])
res_data = reservoir.transform((dataset,))
sum_data = np.sum(res_data, axis=-1)
print(sum_data.shape)


plot_data = sum_data[0, :100]
sns.relplot(kind='line', data=plot_data, legend=False, alpha=0.3)
plt.title('State "Net photosynthesis rate"')
plt.show()
```

```python
from pipeline_hydroshoot import TargetGenerator


target = TargetGenerator(target='output_An', run_ids=[0])
target_data = target.transform((dataset,))
print(target_data.shape)


plot_data = target_data[0, :100]
sns.relplot(kind='line', data=plot_data, legend=False, alpha=0.3)
plt.title('Target "Net photosynthesis rate"')
plt.show()
```
