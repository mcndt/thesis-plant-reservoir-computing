```python
RESULTS_ROOT = 'results/'

experiments = [
  'gdc_can1_grapevine',
  'gdc_can2_grapevine',
  'gdc_can3_grapevine',

  'gdc_can1_grapevine_no_deficit',
  'gdc_can2_grapevine_no_deficit',
  'gdc_can3_grapevine_no_deficit',

  'gdc_can1_grapevine_10_days_no_deficit',
  'gdc_can2_grapevine_10_days_no_deficit',
  'gdc_can3_grapevine_10_days_no_deficit',
]

# notebook = 'visualize_experiment_data.ipynb'
notebook = 'rc_experiments.ipynb'
```
